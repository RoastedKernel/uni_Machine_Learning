# Q1 Inheritance and Polymorphism Relationships

```python
class PC:
    def __init__(self, Id, perfomance_cpu):
        self.Id = Id
        self.perfomance_cpu = perfomance_cpu
        
    def __repr__(self):
        return ("PC ID of "+str(self.Id)+" scored " +str(self.perfomance_cpu)+" in the cpu stress test"+"\n")
    def TurnOn(self):
        print("The PC is turned on")
    
    def TurnOff(self):
        print("The PC is Turned Off")
    
    def __lt__(self,other):
        return self.perfomance_cpu<other.perfomance_cpu
    
    def __eq__(self,other):
        self.perfomance_cpu==other.perfomance_cpu
        
        

        
class HPC(PC):
    def __init__(self, Id, perfomance_cpu, perfomance_gpu):
        
        PC.__init__(self, Id, perfomance_cpu)
        
        self.perfomance_gpu = perfomance_gpu
    
    def __repr__(self):
        return ("HPC ID of "+str(self.Id)+" scored " +str(self.perfomance_cpu)
                +" in the cpu stress test and scored "
                +str(self.perfomance_gpu)+" in the GPU stress test"+"\n")
    def __lt__(self,other):
        return self.perfomance_cpu<other.perfomance_cpu
    
    def __eq__(self,other):
        self.perfomance_cpu==other.perfomance_cpu
    
    #def __lt__(self,other):
     #   return self.perfomance_gpu<other.perfomance_gpu
    
    def TurnOn(self):
        print("The HPC is turned on")
    
    def TurnOff(self):
        print("The HPC is Turned Off")
        
    def Simulation_On(self):
        print("The Simualtion has started")
    
    def Simulation_Off(self):
        print("The Simulation has ended")
    
    
    
class Laptop(PC):
    def __init__(self, Id, perfomance_cpu, Battery):
        PC.__init__(self, Id, perfomance_cpu)
            
        self.Battery = Battery
            
    def __repr__(self):
        
        return ("Laptop ID of "+str(self.Id)+" scored " +str(self.perfomance_cpu)
                +" in the cpu stress test and lasted for "
                +str(self.Battery)+" hours "+"\n")
    
    def __lt__(self,other):
        return self.perfomance_cpu<other.perfomance_cpu
    
    def __eq__(self,other):
        self.perfomance_cpu==other.perfomance_cpu
    
    def TurnOn(self):
        print("The Laptop is turned on")
    
    def TurnOff(self):
        print("The Laptop is Turned Off")
        
    def Battery_Low(Self):
        print("Battery life is low please connect the charger")
    
    def Warning (self):
        print("Battrey is low. system will shutoff")
```

```python

pc_list = []
#pc_list.clear()
for i in range(2,14,3):
    p=PC(i, (i*20))
    h=HPC(i,(i*30),(i*50))
    l=Laptop(i, (i*10), (i-1))
    pc_list.append(p)
    pc_list.append(h)
    pc_list.append(l)
    
for pc in pc_list:
    pc.TurnOn()
    pc.TurnOn()
    if isinstance(pc,HPC):
        pc.Simulation_On()
        pc.Simulation_Off()
    if isinstance(pc,Laptop):
        pc.Battery_Low()
        pc.Warning()
```
