### Object Methods

```python
#class name always follows after the word 'class'
class FinalStudent:
    #initialization function --- it is very important --- 
    def __init__(self,name,degree,courses,age,no_courses):
        #the word 'self' refers to the specific instance of the class
        #class attributes are name, degree, and courses
        self.name=name
        self.degree=degree
        self.courses=courses
        self.age=age
        self.no_courses=no_courses
        
    #printout function - same as the __init__ function, this function exists for all objects, with default values
    #unless otherwise specified, as here
    def __repr__(self):
        return self.name+" enrolled in the degree "+self.degree+" and the courses "+self.courses
    
    
    #functions in a class always have the 'self' parameter as the first parameter - it is a reference to that specific 
    #instance of the object
    def numberOfCourses(self):
        #we split the string based on the ',' delimiter
        allCourses=self.courses.split(",")
        #the total number of courses is the number of different entities
        totalCurrentCourses=len(allCourses)
        #put this in the fuction to return the desired value
        return self.no_courses - totalCurrentCourses
    
    def isEngineering(self):
        return 'Engineering' in self.degree or 'engineering' in self.degree
    
#creating first student
mad=FinalStudent("Mad Scientist", "Engineering Disasters","a, b, c",140,5)
#creating second student
soMad=FinalStudent("Maddest Scientist", "Creating Disasters","a, b, c",140,5)

#adding students to list
student_list=[]
student_list.append(mad)
student_list.append(soMad)

#calling all our defined functions
print(mad.isEngineering())
print(soMad.isEngineering())
print(mad.numberOfCourses())
#printing class information
print(student_list)
print(student_list[0])
```

### Object inheritance

The __Vehicle__ class

```python
#defining the superclass
class Vehicle:
    def __init__(self, speed, doors, passengers, capacity):
        self.speed=speed
        self.doors=doors
        self.passengers=passengers
        self.capacity=capacity

    def __repr__(self):
        return "I am a Vehicle with "+str(self.doors)+" doors. I can carry "+str(self.passengers)+" passengers."
    
    def travel(self):
        print("I am a vehicle and I am travelling. Vroom! Vroom!")
```

The __Tractor__ class

```python
#defining the classes - note that the superclass name follows in brackets after the class name
class Tractor(Vehicle):
    def __init__(self,speed,doors,passengers,capacity,attachments,size):
        #note that we call Vehicle's init method
        Vehicle.__init__(self, speed, doors, passengers, capacity)
        #we only need to focus here on the attributes that are different
        self.attachments=attachments
        self.size=size
    
    def __repr__(self):
        return "I am a beautiful Tractor. My size does not matter as I am beautiful as I am. However, if you must know, I am a size "+self.size+"."
```

The __Car__ class

```python
#defining the classes - note that the superclass name follows in brackets after the class name
class Car(Vehicle):
    def __init__(self,speed,doors,passengers,capacity,fuel_type):
        #note that we call Vehicle's init method
        Vehicle.__init__(self, speed, doors, passengers, capacity)
        #we only need to focus here on the attributes that are different
        self.fuel_type=fuel_type
    
    def __repr__(self):
        return "I am a car. I run on "+self.fuel_type
```

### Multiple Inheritance

The code for the __FlyingVehicle__ class would look like this

```python
#defining another superclass
class FlyingVehicle:
    def __init__(self, number_of_engines, propulsion_mechanism):
        self.engines=number_of_engines
        self.propulsion=propulsion_mechanism

    def __repr__(self):
        return "I am a Flying Vehicle with "+str(self.enginges)+" engines. "
    
    def fly(self):
        print("I am in the air and I fly!")
```

define the __Airplane__ class

```python
#defining multiple inheritance
class Airplane(Vehicle,FlyingVehicle):
    def __init__(self, speed, doors, passengers, capacity, propulsion_mechanism, engines, air_type):

        #we first call the __init__ function of the first superclass
        Vehicle.__init__(self, speed,doors, passengers, capacity)
        
        #then of the second super class
        FlyingVehicle.__init__(self, propulsion_mechanism,engines)

        #then we initialize the rest of the attributes
        self.air_type=air_type
        self.passengers=passengers
        self.capacity=capacity

    def __repr__(self):
        return "I am an Airplane"
    
    #we can call any methods from the superclasses
    def airborne(self):
        FlyingVehicle.fly(self)
    
    def ground(self):
        Vehicle.travel(self)

plane=Airplane(3000,8,3000,30,"jet",2,"commercial")
plane.airborne()
plane.ground()

print(plane)
```

### Practice 

1. Write down the classes for a FightingRobot and CommunicatingRobot.
2. Then write the class for a FightingCommunicatingRobot. You're allowed to define their behaviours, but FightingRobot should have at least a fight function, and CommunicatingRobot should have at least a communicate function.
3. After you define the class hierarchy, make sure you call some functions on the CommunicatingRobot to ensure it works.

### Solution 

```python
#define first the FightingRobot
class FightingRobot:
    def __init__(self, armor, mana):
        self.armor=armor
        self.mana=mana
    
    def fight(self):
        print("I am fighting. My armor is "+str(self.armor)+". My mana is :"+str(self.mana))
 
#then the Communicating Robot
class CommunicatingRobot:
    def __init__(self, comm_range, battery):
        self.comm_range=comm_range
        self.battery=battery
        
    def communicate(self):
        print("I am talking. My battery level is: "+str(self.battery)+". My range is: "+str(self.comm_range))

#the FightingCommunicatingRobot inherits behavior from both superclasses
class FightingCommunicatingRobot(FightingRobot, CommunicatingRobot):
    def __init__(self, armor, mana, comm_range, battery, speed):
        FightingRobot.__init__(self,armor,mana)
        CommunicatingRobot.__init__(self,comm_range, battery)
        self.speed=speed
    
    #we redefine - it's called override - the fight function in the FightingRobot. It first has a communicating stage
    def fight(self):
        #first attempt to communicate
        CommunicatingRobot.communicate(self)

        #then fight
        FightingRobot.fight(self)
        
specialrobot=FightingCommunicatingRobot(100,10,200,75,150)
specialrobot.fight()
```

### Object Composition

The __Heart__ class

```python
class Heart:
    def __init__(self, capacity, health_score):
        self.capacity=capacity
        self.health_score=health_score
    
    def beat(self):
        print("The heart is beating!")
```

The __Human__ class

```python
class Human:
    def __init__(self, heart):
        self.heart=heart

    def run(self):
        print("I am running!")
        heart.beat()
```

Once you define objects for the __Heart__ and __Human__ classes (remember that objects are instances of classes), you can see specific behaviour:

```python
heart=Heart(50,100)
human=Human(heart)
human.run()
```

You can also call the 'beat' function directly on the heart object from within the human:

```python
human.heart.beat()
```

### Interactions between objects using functions
####  Polymorphism

```python
class Fish():
    def swim(self):
        print("The fish is swimming.")

    def swim_backwards(self):
        print("The fish might swim backwards.")

    def skeleton(self):
        print("The fish has a skeleton.")

        
```

Now consider two classes, __Shark__ and __Clownfish__ that inherit from __Fish__ and re-define (or override) those three methods.

```python
class Shark(Fish):
    def swim(self):
        print("The shark is swimming.")

    def swim_backwards(self):
        print("The shark cannot swim backwards, but can sink backwards.")

    def skeleton(self):
        print("The shark's skeleton is made of cartilage.")


class Clownfish(Fish):
    def swim(self):
        print("The clownfish is swimming.")

    def swim_backwards(self):
        print("The clownfish can swim backwards.")

    def skeleton(self):
        print("The clownfish's skeleton is made of bone.")
```

One way in which we can use polymorphism is through object polymorphism. Say we create two objects, one of the __Shark__ class and one for the __Clownfish__ class.


```python
sharkey=Shark()
sharkey.swim()

nemo=Clownfish()
nemo.swim()

print("Showing off polymorphism.")
#here, it does not matter what fish I'm using, as long as the method swim_backwards exists in all classes
for fish in (sharkey,nemo):
    fish.swim_backwards()
```

You can also show off polymorphism in another way, through the definition of a function:

```python
def swimming(fish):
    fish.swim()
    
swimming(sharkey)
swimming(nemo)
```

### Method Resolution Order

```python
class A: 
    def rk(self): 
        print(" In class A") 

class B(A): 
    def rk(self): 
        print(" In class B") 

class C(A): 
    def rk(self): 
        print(" In class C") 

        
# classes ordering 
class D(B, C): 
    def __init__(self): 
        print("Constructor D") 
  
  
# prints the lookup order  
print(D.__mro__) 
print(D.mro()) 
```
